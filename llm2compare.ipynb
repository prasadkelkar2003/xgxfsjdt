```python
import os
import numpy as np
from openai import OpenAI
import requests

# Replace this with your actual Hugging Face token
HF_TOKEN = "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"

# Initialize LLM API client
client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=HF_TOKEN,
)

# Prompt to ask both LLMs
question = "Q: Which planet is known as the Red Planet?\n\nA. Earth\nB. Mars\nC. Jupiter\nD. Venus"

# Two model names
models = {
    "Phi-4": "microsoft/phi-4:nebius",
    "Nemotron": "nvidia/Llama-3.1-Nemotron-70B-Instruct-HF:together"
}

# Step 1: Ask both LLMs
answers = {}
for name, model_id in models.items():
    try:
        response = client.chat.completions.create(
            model=model_id,
            messages=[{"role": "user", "content": question}]
        )
        ans = response.choices[0].message.content.strip()
        answers[name] = ans
    except Exception as e:
        answers[name] = f"Error: {e}"

# Step 2: Embed the answers using Hugging Face API
def get_embedding(text):
    url = "https://api-inference.huggingface.co/pipeline/feature-extraction/sentence-transformers/all-MiniLM-L6-v2"
    headers = {"Authorization": f"Bearer {HF_TOKEN}"}
    response = requests.post(url, headers=headers, json={"inputs": text})
    response.raise_for_status()
    return np.array(response.json()[0])

# Get embeddings
embedding_1 = get_embedding(list(answers.values())[0])
embedding_2 = get_embedding(list(answers.values())[1])

# Step 3: Compute cosine similarity
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

similarity = cosine_similarity(embedding_1, embedding_2)

# Step 4: Decide based on similarity
print("\nüîç Comparison of Answers:\n")
if similarity > 0.9:
    print(f"‚úÖ Both LLMs agree (similarity: {similarity:.2f}):\n\n{list(answers.values())[0]}")
else:
    print(f"‚ö†Ô∏è LLMs differ (similarity: {similarity:.2f}):\n")
    for name, answer in answers.items():
        print(f"{name} ‚ûú {answer}")
```

```python
import re
from huggingface_hub import InferenceClient
from openai import OpenAI

# Set your Hugging Face Inference API key
HF_TOKEN = "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"

# === Question to ask ===
prompt = "Q: What gas do plants absorb from the atmosphere for photosynthesis?\n\nA. Oxygen\nB. Nitrogen\nC. Carbon Dioxide\nD. Hydrogen"

# === Define LLM models to query ===
models = {
    "Phi-4": "microsoft/phi-4:nebius",
    "Nemotron": "nvidia/Llama-3.1-Nemotron-70B-Instruct-HF:together"
}

# === Initialize clients ===
llm_client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=HF_TOKEN
)

# === Function to query LLM ===
def query_llm(model_id, question):
    try:
        response = llm_client.chat.completions.create(
            model=model_id,
            messages=[ {"role": "user", "content": question}]
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        return f"Error: {e}"

# === Extract MCQ answer (e.g., "B. Mars") ===
def extract_answer(text):
    match = re.search(r"\b([A-D])[\.\)]\s*([^\n]*)", text)
    if match:
        return f"{match.group(1)}. {match.group(2).strip()}"
    return text.strip()[:50]

# === Function to compare using another LLM (Zephyr) ===
def judge_similarity_with_llm(ans1, ans2):
    judgment_prompt = f"""
Given the two multiple choice answers below, determine if they are the same answer (ignore explanations or additional text).

Answer 1: {ans1}
Answer 2: {ans2}

Respond with "Yes" if they refer to the same answer choice (like both are "B. Mars"). Otherwise, respond with "No".
    """.strip()

    try:
        response = llm_client.chat.completions.create(
            model="HuggingFaceH4/zephyr-7b-beta:featherless-ai",
            messages=[{"role": "user", "content": judgment_prompt}]
        )
        decision = response.choices[0].message.content.strip().lower()
        return "yes" in decision
    except Exception as e:
        print(f"Zephyr comparison failed: {e}")
        return False

# === Main Execution ===
answers = {}
stripped_answers = {}

print("üîç LLM Answer Comparison:\n")

for name, model_id in models.items():
    raw = query_llm(model_id, prompt)
    parsed = extract_answer(raw)
    answers[name] = raw
    stripped_answers[name] = parsed
    print(f"{name} ‚ûú {parsed}")

# === Compare using Zephyr judgment ===
vals = list(stripped_answers.values())
same_answer = judge_similarity_with_llm(vals[0], vals[1])

print("\n‚úÖ Final Verdict:\n")
if same_answer:
    print("‚úÖ Both LLMs gave similar answers (according to Zephyr)")
    print(f"\nüü¢ Final Answer: {vals[0]}")
else:
    print("‚ö†Ô∏è LLMs gave different answers according to Zephyr judgment.")
    for name, ans in stripped_answers.items():
        print(f"{name} ‚ûú {ans}")
```
